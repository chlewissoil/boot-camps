###Very Basic Plotting

We make some data to plot:
        


```python
import numpy as n
demo = n.random.random((5,4))
demo[0,0] = -0.1
demo[-1,-1] = 1.1
demo
```

Slicing the array (a list of lists):

```python
demo[0,:], demo[:,0]
```

And plot it:

```python
import matplotlib.pyplot as plt

plt.title('Demo data')
plt.plot(demo)
```

EXERCISE: that's four lines with five data points each: why?

Answer: We have four columns with five elements each. Each column is used as a data series. We added -0.1 and 1.1 values to make it easy to pick out those columns. 

What if my series are in rows? Numpy arrays have a lot of built-in reorganizing, and switching rows and arrays is an example:

```python

plt.title('Demo transposed')
plt.plot(demo.T)
```

EXERCISE: Look at the x-axis. Why those values?

Answer: In both cases, `plt.plot` assumed that the data is categorical and plotted each value against (0,1,2,3,4) and (0,1,2,3) respectively. 

Maybe the first (0th) column should be the x-axis:

```python

plt.xlabel('Independent variable')
plt.plot(demo[:,0], demo[:,1:])
```

### Self-Describing Plots (and Data)

This needs line labels. What do the columns mean? That should be built into our data, like column headers in a CSV file. You can add column names to an array of data and create a record array, a _recarray_:


```python
demo_rec = demo.view(n.recarray)
demo_rec.dtype=[('x', '<f8'), ('a', '<f8'), ('b', '<f8'), ('c', '<f8')]
demo_rec.x                       
```

*Warning*: The data is still shared between `demo_rec` and `demo`:

```python
demo_rec.x[0] = -0.2
demo[:,0]
```

# Writing your own functions

If you're cutting and pasting code, or if your loops are so full they're hard to read, pull out a function. Redoing the fitting example this way:

```python
def plot_and_fit(independent, dependent, name):
    # Lines? Look at the output from the very first plot  
    lines = plt.plot(independent, dependent, marker='o',
                     linestyle='none', label=name)
    #fit a line
    coeff = n.polyfit(independent.flatten(), 
                      dependent.flatten(), 1)
    fit_y = n.polyval(coeff, independent)
    plt.plot(independent, fit_y, linestyle=':', 
             color = lines[0].get_color(),
             label = name + ' fit')
    

for col in demo_rec.dtype.names:
    if col is not 'x': 
        plot_and_fit(demo_rec.x, demo_rec[col], col)
plt.legend()
 
```
